[Securities Available for Equity Segments](https://www.nseindia.com/market-data/securities-available-for-trading)

```python
# Read Equities Names
equity_details = pd.read_csv('EQUITY_L.csv')
```

```python
equity_details.SYMBOL
```

```python
# Download Data for 5 NS
for name in equity_details.SYMBOL[:5]:
    data = yf.download(f'{name}.NS')
    data.to_csv(f'./Data/{name}.csv')
    
```

```python
# Download data for all list

for name in equity_details.SYMBOL:
    try:
        data = yf.download(f'{name}.NS')
        data.to_csv(f'./Data/{name}.csv')
    except Exceptions as e:
        print(f'{name}===> {e}')
```
